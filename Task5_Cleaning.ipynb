```python
import pandas as pd
```

```python
file_name = "titanic_data.csv"
pd.read_csv(file_name)
```

```python
df = pd.read_csv(file_name)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
missing = df.isnull().sum()
columns_to_clean = missing[missing > 0]
columns_to_clean
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
rows_before = df.shape[0]
rows_before
```

```python
df.drop_duplicates(inplace=True)
```

```python
rows_after = df.shape[0]
rows_after
```

```python
df.dtypes
```

```python
df['Survived'] = df['Survived'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
df['Sex'] = df['Sex'].astype('category')
df['Embarked'] = df['Embarked'].astype('category')
```

```python
df.dtypes
```

```python
df['Fare_Band'] = pd.cut(
    df['Fare'],
    bins=[0, 10, 30, 100, 600],
    labels=['Low', 'Medium', 'High', 'Very High']
)

df[['Fare', 'Fare_Band']].head()
```

```python
def age_category(age):
    if age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

df['Age_Category'] = df['Age'].apply(age_category)

df[['Age', 'Age_Category']].head()
```

**Titanic Dataset:**
Data Cleaning Summary

**1. Identifying Missing Values**

* Missing values were checked using `.isnull().sum()`.
* The Age column had 177 missing values, Embarked had 2 missing values, and Cabin had 687 missing values.
* Columns with missing values were analyzed to decide whether to fill or drop them before modeling or reporting.

**2. Handling Missing Values**

* Age missing values were filled using the **median** to reduce the effect of outliers.
* Embarked missing values were filled using the **mode** since it is categorical.
* Cabin was **dropped** due to too many missing entries.

**3. Removing Duplicates**

* Duplicate rows were checked and removed to ensure the dataset remained accurate and clean.

**4. Correcting Data Types**

* Columns like Survived, Pclass, Sex, and Embarked were converted to **category** type for better analysis.
* No date columns were present, so datetime conversion was not needed.

**5. Feature Engineering**

* A new column, **Age_Category**, was created to group passengers into Child, Adult, and Senior.
* Another column, **Fare_Band**, was created to classify ticket prices as Low, Medium, High, or Very High.
